Сбор данных в один DataFrame

```python
def load_all_csv(zip_path):
    with tempfile.TemporaryDirectory() as temp_dir:  # Временная папка для распаковки
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(temp_dir)  # Распаковываем ZIP
        directory = temp_dir  # Путь к распакованным файлам
        
        dfs = []
        for root, dirs, files in os.walk(directory):  # Рекурсивный обход всех подпапок (например, 'data')
            for filename in files:
                if filename.endswith('.json'):  # Ищем .json с CSV-содержимым
                    filepath = os.path.join(root, filename)
                    try:
                        df = pd.read_csv(filepath, low_memory=False)  # Читаем как CSV
                        df['original_file'] = filename  # Отслеживаем источник (кафе, парки и т.д.)
                        dfs.append(df)
                    except pd.errors.ParserError:
                        print(f"Ошибка парсинга файла {filename} — проверьте формат (возможно, не CSV внутри).")
        if not dfs:
            raise ValueError("Нет файлов для объединения — проверьте содержимое ZIP (ожидаются .json с CSV-структурой в папках).")
        combined_df = pd.concat(dfs, ignore_index=True)
        print(f"Объединено {len(dfs)} файлов, всего строк: {len(combined_df)}")
    return combined_df  # Временная директория удалится автоматически
```

Удаление дубликатов

```python
def remove_duplicates(df):
    df['unique_key'] = (
        df['title'].str.lower() + '_' +
        df['address'].str.lower() + '_' +
        df['latitude'].astype(str) + '_' +
        df['longitude'].astype(str)
    )
    df = df.sort_values(by=['unique_key', 'review_count'], ascending=[True, False])
    df = df.drop_duplicates(subset='unique_key', keep='first')  # С max отзывами для рекомендаций
    df = df.drop(columns=['unique_key'])
    print(f"После удаления дубликатов: {len(df)} строк")
    return df
```

```python
df_clean = remove_duplicates(df)
```

Добавление фичей из столбца about

```python
df_clean['about'].isna().sum()
```

```python
# Функция для парсинка json из строк
def safe_parse_json(json_str):
    if pd.isna(json_str) or not isinstance(json_str, str) or json_str.strip() == '':
        return None
    try:
        return json.loads(json_str)
    except json.JSONDecodeError as e:
        print(f"Ошибка парсинга JSON в столбце 'about': {e}")
        return None
```

```python
# Собираем все уникальные имена опций из столбца 'about'
unique_options = set()
for about_str in df_clean['about']:
    about_data = safe_parse_json(about_str)
    if about_data:
        for category in about_data:
            options = category.get('options', [])
            for option in options:
                option_name = option.get('name', '')
                if option_name:
                    # Преобразуем имя опции в безопасное для столбца (заменяем пробелы на подчеркивания, убираем спецсимволы)
                    safe_name = option_name.lower().replace(' ', '_').replace('+', '_plus').replace('-', '_')
                    unique_options.add(safe_name)
```

```python
# Функция для извлечения фич для одной строки
def extract_about_features(row):
    about_data = safe_parse_json(row['about'])
    # Инициализируем словарь с np.nan для всех возможных опций
    features = {option: np.nan for option in unique_options}
    
    if about_data:
        for category in about_data:
            options = category.get('options', [])
            for option in options:
                option_name = option.get('name', '')
                if option_name:
                    safe_name = option_name.lower().replace(' ', '_').replace('+', '_plus').replace('-', '_')
                    features[safe_name] = option.get('enabled', False)
    
    return pd.Series(features)
```

```python
# Применяем функцию и добавляем новые столбцы в df_clean
about_features = df_clean.apply(extract_about_features, axis=1)
df_clean = pd.concat([df_clean, about_features], axis=1)
```
